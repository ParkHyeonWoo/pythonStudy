```python
tmp_df = pd.DataFrame(tmp, columns = ['TEAMPOSITION', 'SUMMONER1ID', 'SUMMONER2ID'])
tmp_df
```

```python
tmp_df.rename(columns={'SUMMONER1ID':'D_spell', 'SUMMONER2ID':'F_spell'}, inplace=True)
tmp_df
```

```python
tmp_df['D_spell'] =  tmp_df.apply(lambda x : 1 if x['D_spell'] == '4' else 0, axis = 1)
```

```python
tmp_df['F_spell'] =  tmp_df.apply(lambda x : 1 if x['F_spell'] == '4' else 0, axis = 1)
```

```python
tmp_df
```

```python
group_df = tmp_df.groupby('TEAMPOSITION').sum()
group_df
```

```python
sum_df = pd.DataFrame([group_df.sum()], index = ['all'])
sum_df
```

```python
all_df=group_df.append(sum_df)
all_df
```

```python
plt.figure()
plt.pie(all_df.loc['TOP'], labels = ['D_spell','F_spell'], shadow = False, autopct = '%.2f%%',
       colors = ['lightgreen', 'lightblue'], explode = (0,0.1))
plt.grid(True)
plt.legend(loc = 4)
plt.title('탑 D,F 점멸 선호도')
```
