```python
def calc(num, cal):
    if cal == '+':
        num = num + 0.123
    elif cal == '-':
        num = num - 0.123
    else:
        return False
    return num
        
calc(3, '+')
```

```python
champion_kr_name_df.rename(columns={'key':'championId'}, inplace = True)
champion_kr_name_df
```

```python
lol_df = pd.merge(data_solo_rank, champion_kr_name_df, on=['championId'])
lol_df
```

```python
lol_df[['gameId', 'name', 'duration', 'version', 'participantId', 'championId', 'lane',
       'side', 'win', 'damageDealt', 'damageTaken', 'gold', 'kills', 'deaths',
       'assists']]
```

```python
lol_df['kda'] = 'none'
lol_df
```

```python
def kda(k, d, a):
    try:
        kda = round((k+a)/d, 2)
        return kda
    except:
        kda = 'perfect'
        return kda
```

```python
lol_df['kda'] = lol_df.apply(lambda x: kda(x.kills, x.deaths, x.assists), axis = 1)
```
