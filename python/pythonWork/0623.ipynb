```python
import pandas as pd
import requests
```

```python
url = "https://api.upbit.com/v1/candles/days"

querystring = {"market":"KRW-ETH","count":"200"}

headers = {"Accept": "application/json"}

response = requests.request("GET", url, headers=headers, params=querystring)
```

```python
res = response.json()
```

```python
df = pd.DataFrame(res)
```

```python
my_df = df[['market','candle_date_time_kst','opening_price','high_price','low_price','trade_price']].copy()
my_df
```

```python
my_df['candle_date_time_kst'] = my_df['candle_date_time_kst'].str[:10]
my_df
```

```python
my_df.set_index('candle_date_time_kst', inplace = True) #지정한 컬럼을 인덱스로 변경한다.
```

```python
my_df
```

```python
my_df['m5'] = 0
my_df['m10'] = 0
```

```python
print(my_df.trade_price[:5].mean())
print(my_df.trade_price[:10].mean())
```

```python
def df_mean(x):
    for i in range(len(x)):
        x['m5'].iloc[i] = x.iloc[i:i+5].trade_price.mean()
        x['m10'].iloc[i] = x.iloc[i:i+10].trade_price.mean()
        if len(x.iloc[i:i+5]) != 5:
            x['m5'].iloc[i] = 0
        if len(x.iloc[i:i+10]) != 10:
            x['m10'].iloc[i] = 0
```

```python
df_mean(my_df)
```
