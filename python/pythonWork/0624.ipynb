아나콘다 프롬포트 -> conda install tqdm, pip install tqdm

```python
from tqdm import tqdm
```

```python
for i in tqdm(range(len(my_df))):    
    query3 = (
    f"insert into housePrice(자치구명, 법정동명, 지번구분명, 층수, 계약일, 전월세, 면적, 보증금, 임대료, 건물명, 건축년도)"
    f"values(\'{my_df.자치구명.iloc[i]}\', \'{my_df.법정동명.iloc[i]}\', \'{my_df.지번구분명.iloc[i]}\', \'{my_df.층수.iloc[i]}\',"
    f"\'{my_df.계약일.iloc[i]}\', \'{my_df.전월세.iloc[i]}\', {my_df.면적.iloc[i]}, {my_df.보증금.iloc[i]},"
    f"{my_df.임대료.iloc[i]}, \'{my_df.건물명.iloc[i]}\', \'{my_df.건축년도.iloc[i]}\')"
    )
    my_utils.sql_execute(query3)
```

```python
my_utils.db_close()
```

```python
def insert(my_df):   
    query4 = (
    f"insert into housePrice(자치구명, 법정동명, 지번구분명, 층수, 계약일, 전월세, 면적, 보증금, 임대료, 건물명, 건축년도)"
    f"values(\'{my_df.자치구명}\', \'{my_df.법정동명}\', \'{my_df.지번구분명}\', \'{my_df.층수}\',"
    f"\'{my_df.계약일}\', \'{my_df.전월세}\', {my_df.면적}, {my_df.보증금},"
    f"{my_df.임대료}, \'{my_df.건물명}\', \'{my_df.건축년도}\')"
    )
    my_utils.sql_execute(query4)
    return
```

```python
tqdm.pandas()
my_df.progress_apply(lambda x: insert(x), axis = 1)
```
