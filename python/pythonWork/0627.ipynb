```python
def summonerName(name):
    url = 'https://kr.api.riotgames.com/lol/summoner/v4/summoners/by-name/'+name+'?api_key='+api_key
    res = requests.get(url).json()
    return res['puuid']
```

```python
uid = summonerName('hideonbush')
uid
```

```python
def puuid(uid, num):
    url= 'https://asia.api.riotgames.com/lol/match/v5/matches/by-puuid/'+uid+'/ids?start=0&count='+str(num)+'&api_key='+api_key
    res = requests.get(url).json()
    return res
```

```python
match = puuid(uid, 5)
match
```

```python
def match_Up(match_Number):
    mat = 'https://asia.api.riotgames.com/lol/match/v5/matches/'+match_Number+'?api_key='+api_key
    res_mat = requests.get(mat).json()
    mat_ti = 'https://asia.api.riotgames.com/lol/match/v5/matches/'+match_Number+'/timeline?api_key='+api_key
    res_mat_ti = requests.get(mat_ti).json()
    return res_mat, res_mat_ti
```

```python
def summoner(s_Name):
    tmp = []
    for i in tqdm(s_Name[:5]):
        uid = summonerName(i)
        match = puuid(uid, 5)
        matchUp, matchTime = match_Up(match[0])
        tmp.append([match[0], matchUp, matchTime])
    return tmp
```

```python
def grand_lst(df):
    lst = []
    for i in tqdm(range(len(df))):
        tmp = []
        for j in tqdm(range(len(df.iloc[i].matchs['info']['participants']))):
            if df.iloc[i].matchs['info']['participants'][j]['deaths'] != 0:
                tmp.append([df.iloc[i].matchs['info']['gameId'], df.iloc[i].matchs['info']['participants'][j]['championName'], df.iloc[i].matchs['info']['participants'][j]['summonerName'], round(df.iloc[i].matchs['info']['participants'][j]['kills']+df.iloc[i].matchs['info']['participants'][j]['assists']/df.iloc[i].matchs['info']['participants'][j]['deaths'], 2)])
            else:
                tmp.append([df.iloc[i].matchs['info']['gameId'], df.iloc[i].matchs['info']['participants'][j]['championName'], df.iloc[i].matchs['info']['participants'][j]['summonerName'], 'perfect'])
        lst.extend(tmp)
    return lst
```
