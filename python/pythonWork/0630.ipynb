```python
my_df = df[df.GAMEID == 'KR_5993841747'].sort_values(by=['PARTICIPANTID'])
```

```python
my_df = my_df[['GAMEID','SUMMONERNAME','CHAMPIONNAME','TEAMPOSITION','TEAMID','WIN','KILLS','DEATHS','ASSISTS','G_14']]
```

```python
my_df
```

```python
blue_team = my_df[my_df.TEAMID == 100]
red_team = my_df[my_df.TEAMID == 200].drop(columns=['TEAMID','WIN']).rename(columns = {'SUMMONERNAME':'red_SUMMONERNAME','CHAMPIONNAME':'red_CHAMPIONNAME','KILLS':'red_KILLS','DEATHS':'red_DEATHS','ASSISTS':'red_ASSISTS','G_14':'red_G_14'})
```

```python
blue_team.sort_values(by = ['G_14'],ascending = False)
```

```python
red_team.sort_values(by = ['red_G_14'],ascending = False)
```

```python
game_stat_df = pd.merge(blue_team,red_team,on=['GAMEID','TEAMPOSITION'])
```

```python
game_stat_df['lane_win'] = 'blue_win' or 'red_win'
```

```python
def win(x):
    if x.G_14 - x.red_G_14 > 0 :
        return 'blue_win'
    elif x.G_14 - x.red_G_14 == 0:
        return 'draw'
    else:
        return 'red_win'
```

```python
game_stat_df['lane_win'] = game_stat_df.apply(lambda x:win(x),axis=1)
```

```python
game_stat_df
```

```python
my_df2 = df[['GAMEID','CHAMPIONNAME','TEAMID','TEAMPOSITION','WIN','G_14']]
```

```python
my_df2['WIN'] = my_df2.apply(lambda x:1 if x.WIN == 'True' else 0 ,axis =1)
```

```python
my_df2
```

```python
cnt_df = my_df2.groupby(['CHAMPIONNAME','TEAMPOSITION'])[['WIN']].count().rename(columns = {'WIN':'CNT'})
```

```python
win_df = my_df2.groupby(['CHAMPIONNAME','TEAMPOSITION'])[['WIN']].sum()
```

```python
cnt_df
```

```python
cnt_win_df = pd.merge(cnt_df,win_df,on = ['CHAMPIONNAME','TEAMPOSITION'])
```

```python
cnt_win_df['win_rate'] = round(cnt_win_df.WIN / cnt_win_df.CNT,4)
```

```python
cnt_win_df.sort_values(by = ['win_rate'],ascending = False)[:60]
```
