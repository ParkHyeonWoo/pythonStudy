```python
df_tmp['JCG_DT'].str.split('.', expand = True).rename(columns = {0:'년',1:'월',2:'일',3:''})
```

```python
def preprocessing(x):
    x_list = x.split('.')
    res = ''.join(x_list[:3])
    return int(res)
```

```python
df_tmp['JCG_DT'] = df_tmp.apply(lambda x: preprocessing(x.JCG_DT), axis = 1)
```

```python
df_tmp = df_tmp[df_tmp.JCG_DT > 20000000]
```

```python
def bar_stack(data, rotation, title, ylim = None, stacked = False, yticks_interval = 50000):
    plt.figure()
    data.plot(kind = 'bar', rot = rotation, title = title, legend = True, stacked = stacked)
    plt.legend(loc = 'best')
    
    if stacked == False:
        maxlim = (int(max(data.max()/yticks_interval)+1)) * yticks_interval
        values = np.arange(0, maxlim + 1, yticks_interval)
        plt.yticks(values, ['%s'%format(val, ',') for val in values])
    else:
        maxlim = (int(max(data.max(axis = 1)/yticks_interval)+1)) * yticks_interval
        values = np.arange(0, maxlim + 1, yticks_interval)
        plt.yticks(values, ['%s'%format(val, ',') for val in values])
```

```python
sample_df = sample_df.astype({'GANGNAM':'int', 'GANGNAMADD':'int', 'GANGDONG':'int','GANGDONGADD':'int'})
```

```python
bar_stack(data = sample_df, rotation = 45, title = '지역별 코로나 확진자 동향')
```
