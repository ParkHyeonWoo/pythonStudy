```python
match['first_blood'] = match.apply(lambda x: 1 if (str(x.PARTICIPANTID) == x.KILLERID.split('|')[0]) else 0, axis = 1)
```

```python
def point_k(t, k, a, log):
    tmp = log.split('|')
    if t == 100:
        cnt = list(filter(lambda x: int(x)<=5, tmp))
    else:
        cnt = list(filter(lambda x: int(x)>5, tmp))
    try:
        return round((k + a) / len(cnt), 3)
    except:
        return 0
```

```python
match['kill_point'] = match.apply(lambda x: point_k(x.TEAMID ,x.KILLS, x.ASSISTS, x.KILLERID), axis = 1)
```

```python
tmp_all_kills = match[['GAMEID', 'TEAMID', 'KILLS']].groupby(['GAMEID', 'TEAMID']).sum().\
                rename(columns = {'KILLS':'all_kills'})
```

```python
tmp_all_kills
```

```python
match_tmp = pd.merge(match, tmp_all_kills, on = ['GAMEID', 'TEAMID'])
```

```python
match_tmp['kill_point'] = round((match_tmp['KILLS'] + match_tmp['ASSISTS'])/match_tmp['all_kills'], 3)
```

```python
match['WIN'] = match.apply(lambda x: 1 if (x.WIN == 'True') else 0, axis = 1)
```
