```python
df['WIN'] = df.apply(lambda x: 1 if (x.WIN == 'True') else 0, axis = 1)
```

```python
blue_top = df[(df.TEAMID == '100') & (df.TEAMPOSITION == 'TOP')][['GAMEID', 'CHAMPIONNAME', 'WIN']].\
            rename(columns = {'CHAMPIONNAME':'t_champ'})
blue_jg = df[(df.TEAMID == '100') & (df.TEAMPOSITION == 'JUNGLE')][['GAMEID', 'CHAMPIONNAME', 'WIN']].\
            rename(columns = {'CHAMPIONNAME':'j_champ'})
```

```python
red_top = df[(df.TEAMID == '200') & (df.TEAMPOSITION == 'TOP')][['GAMEID', 'CHAMPIONNAME', 'WIN']].\
            rename(columns = {'CHAMPIONNAME':'t_champ'})
red_jg = df[(df.TEAMID == '200') & (df.TEAMPOSITION == 'JUNGLE')][['GAMEID', 'CHAMPIONNAME', 'WIN']].\
            rename(columns = {'CHAMPIONNAME':'j_champ'})
```

```python
blue_team = pd.merge(blue_top, blue_jg, on = ['GAMEID', 'WIN'])
```

```python
red_team = pd.merge(red_top, red_jg, on = ['GAMEID', 'WIN'])
```

```python
blue_team
```

```python
blue_tmp = blue_team[['GAMEID', 't_champ', 'j_champ']].\
            rename(columns = {'t_champ':'enemy_t_champ', 'j_champ':'enemy_j_champ'})
red_tmp = red_team[['GAMEID', 't_champ', 'j_champ']].\
            rename(columns = {'t_champ':'enemy_t_champ', 'j_champ':'enemy_j_champ'})
```

```python
blue_tmp
```

```python
tmp = pd.merge(blue_team, red_tmp, on = ['GAMEID']).append(pd.merge(red_team, blue_tmp, on = ['GAMEID']))
```

```python
tmp
```

```python
games = tmp[['t_champ', 'j_champ', 'enemy_t_champ', 'enemy_j_champ', 'WIN']].\
        groupby(['t_champ', 'j_champ', 'enemy_t_champ', 'enemy_j_champ']).count().rename(columns = {'WIN':'games'})
```

```python
games
```

```python
win_cnt = tmp[['t_champ', 'j_champ', 'enemy_t_champ', 'enemy_j_champ', 'WIN']].\
        groupby(['t_champ', 'j_champ', 'enemy_t_champ', 'enemy_j_champ']).sum().rename(columns = {'WIN':'win_cnt'})
```

```python
win_cnt
```

```python
result = games.join(win_cnt)
```

```python
result['win_rate'] = round(result.win_cnt/result.games, 3)
```

```python
result.sort_values(by = ['games'], ascending = False)
```

```python
blue_bot = df[(df.TEAMID == '100') & (df.TEAMPOSITION == 'BOTTOM')][['GAMEID', 'CHAMPIONNAME', 'WIN', 'G15']].\
            rename(columns = {'CHAMPIONNAME':'b_champ'})
blue_sup = df[(df.TEAMID == '100') & (df.TEAMPOSITION == 'UTILITY')][['GAMEID', 'CHAMPIONNAME', 'WIN']].\
            rename(columns = {'CHAMPIONNAME':'s_champ'})
blue_team = pd.merge(blue_bot, blue_sup, on = ['GAMEID', 'WIN'])
```

```python
blue_team
```

```python
red_bot = df[(df.TEAMID == '200') & (df.TEAMPOSITION == 'BOTTOM')][['GAMEID', 'CHAMPIONNAME', 'WIN', 'G15']].\
            rename(columns = {'CHAMPIONNAME':'b_champ'})
red_sup = df[(df.TEAMID == '200') & (df.TEAMPOSITION == 'UTILITY')][['GAMEID', 'CHAMPIONNAME', 'WIN']].\
            rename(columns = {'CHAMPIONNAME':'s_champ'})
red_team = pd.merge(red_bot, red_sup, on = ['GAMEID', 'WIN'])
```

```python
blue_tmp = blue_team[['GAMEID', 'b_champ', 's_champ', 'G15']].\
            rename(columns = {'b_champ':'enemy_b_champ', 's_champ':'enemy_s_champ', 'G15':'enemy_g15'})
red_tmp = red_team[['GAMEID', 'b_champ', 's_champ', 'G15']].\
            rename(columns = {'b_champ':'enemy_b_champ', 's_champ':'enemy_s_champ', 'G15':'enemy_g15'})
```

```python
tmp = pd.merge(blue_team, red_tmp, on = ['GAMEID']).append(pd.merge(red_team, blue_tmp, on = ['GAMEID']))
```

```python
tmp
```

```python
tmp['lane_win'] = tmp.apply(lambda x: 1 if x.G15 > x.enemy_g15 else 0, axis = 1)
```

```python
games = tmp[['b_champ', 's_champ', 'enemy_b_champ', 'enemy_s_champ', 'WIN']].\
        groupby(['b_champ', 's_champ', 'enemy_b_champ', 'enemy_s_champ']).count().rename(columns = {'WIN':'games'})
win_cnt = tmp[['b_champ', 's_champ', 'enemy_b_champ', 'enemy_s_champ', 'WIN']].\
        groupby(['b_champ', 's_champ', 'enemy_b_champ', 'enemy_s_champ']).sum().rename(columns = {'WIN':'win_cnt'})
lane_win_cnt = tmp[['b_champ', 's_champ', 'enemy_b_champ', 'enemy_s_champ', 'lane_win']].\
        groupby(['b_champ', 's_champ', 'enemy_b_champ', 'enemy_s_champ']).sum().rename(columns = {'lane_win':'lane_win_cnt'})
```

```python
result = games.join(win_cnt).join(lane_win_cnt)
```

```python
result['win_rate'] = round(result.win_cnt/result.games, 3)
result['lane_win_rate'] = round(result.lane_win_cnt/result.games, 3)
```

```python
result.sort_values(by = ['games'], ascending = False)
```
